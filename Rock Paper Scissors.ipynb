```python
import random
import tkinter as tk
from tkinter import messagebox

def get_computer_choice():
    return random.choice(['rock', 'paper', 'scissors'])

def determine_winner(user, computer):
    if user == computer:
        return 'tie'
    elif (user == 'rock' and computer == 'scissors') or \
         (user == 'scissors' and computer == 'paper') or \
         (user == 'paper' and computer == 'rock'):
        return 'user'
    else:
        return 'computer'

def play(user_choice):
    global user_score, computer_score
    computer_choice = get_computer_choice()
    result = determine_winner(user_choice, computer_choice)
    
    if result == 'tie':
        result_text = f"It's a tie! You both chose {user_choice}."
        result_label.config(text=result_text, fg='blue')
    elif result == 'user':
        result_text = f"You win! {user_choice} beats {computer_choice}."
        result_label.config(text=result_text, fg='green')
        user_score += 1
    else:
        result_text = f"Computer wins! {computer_choice} beats {user_choice}."
        result_label.config(text=result_text, fg='red')
    
    score_label.config(text=f"Score -> User: {user_score}, Computer: {computer_score}")

def quit_game():
    messagebox.showinfo("Game Over", f"Final Scores -> User: {user_score}, Computer: {computer_score}")
    root.destroy()

# GUI Setup
root = tk.Tk()
root.title("Rock-Paper-Scissors")
root.geometry("400x400")
root.config(bg='#f0f8ff')
user_score = 0
computer_score = 0

title_label = tk.Label(root, text="Rock-Paper-Scissors", font=("Arial", 24, 'bold'), bg='#f0f8ff', fg='#4b0082')
title_label.pack(pady=10)

result_label = tk.Label(root, text="Make your choice!", font=("Arial", 16), bg='#f0f8ff', fg='#00008b')
result_label.pack(pady=10)

score_label = tk.Label(root, text="Score -> User: 0, Computer: 0", font=("Arial", 14), bg='#f0f8ff', fg='#8b0000')
score_label.pack(pady=10)

button_frame = tk.Frame(root, bg='#f0f8ff')
button_frame.pack(pady=20)

rock_button = tk.Button(button_frame, text="Rock", command=lambda: play('rock'), font=("Arial", 12), bg='#ff6347', fg='white', width=10)
rock_button.grid(row=0, column=0, padx=10, pady=5)

paper_button = tk.Button(button_frame, text="Paper", command=lambda: play('paper'), font=("Arial", 12), bg='#4682b4', fg='white', width=10)
paper_button.grid(row=0, column=1, padx=10, pady=5)

scissors_button = tk.Button(button_frame, text="Scissors", command=lambda: play('scissors'), font=("Arial", 12), bg='#32cd32', fg='white', width=10)
scissors_button.grid(row=0, column=2, padx=10, pady=5)

quit_button = tk.Button(root, text="Quit", command=quit_game, font=("Arial", 12), bg='#8b0000', fg='white', width=10)
quit_button.pack(pady=20)

root.mainloop()
```
